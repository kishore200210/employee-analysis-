```python
import pandas as pd
```

```python
file=pd.read_csv(r"C:\Users\ADMIN\OneDrive\Desktop\employe_project\Employee-Attrition - Employee-Attrition.csv")
df=pd.DataFrame(file)
df.head(3)
```

```python
#Remove duplicate
df = df.loc[:, ~df.columns.duplicated()]
df.head(3)
```

```python
#outlier values
df = df.applymap(lambda x: None if isinstance(x,(int,float)) and x < 0 else x)
```

```python
#Removing Uninformative Variables
df.drop(["EmployeeCount", "EmployeeNumber"], axis=1, inplace=True)
```

```python
#Encoding Categorical Variables
df = pd.get_dummies(df, drop_first=True)
```

```python
#Scaling 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numeric_cols = ['Age','DailyRate','DistanceFromHome']
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```

```python
#converting cleaned file to csv
df.to_csv("cleaned_file.csv")
```
