```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("cleaned_file.csv")
df.head(3)
```

```python
# Dataset shape
df.shape

# Column names
df.columns

# Basic info
df.info()

# Missing values
df.isnull().sum()
```

```python
df.describe()
```

```python
#Target Variable Attrition Distribution
plt.figure(figsize=(7,4))
sns.countplot(data=df, x='Attrition_Yes')
plt.title("Attrition Distribution")
plt.show()

attrition_rate = df['Attrition_Yes'].mean()*100
print(f"Employee Attrition Rate: {attrition_rate:.2f}%")
```

```python
#Age Distribution
plt.figure(figsize=(8,4))
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
#Attrition by Age
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Attrition_Yes', y='Age')
plt.title("Attrition by Age")
plt.show()
```

```python
#Attrition by Monthly Income
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Attrition_Yes', y='MonthlyIncome')
plt.title("Attrition vs Monthly Income")
plt.show()
```

```python
#Attrition by Distance From Home
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Attrition_Yes', y='DistanceFromHome')
plt.title("Attrition vs Distance From Home")
plt.show()
```

```python
#Work-Life Balance vs Attrition
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='WorkLifeBalance', hue='Attrition_Yes')
plt.title("WorkLifeBalance vs Attrition")
plt.show()
```

```python
#Job Satisfaction vs Attrition
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='JobSatisfaction', hue='Attrition_Yes')
plt.title("JobSatisfaction vs Attrition")
plt.show()
```

```python
#Environment Satisfaction vs Attrition

plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Attrition_Yes', y='EnvironmentSatisfaction', estimator=np.mean, ci=95)
plt.title("Average Environment Satisfaction by Attrition")
plt.xlabel("Attrition (0 = No, 1 = Yes)")
plt.ylabel("Mean Environment Satisfaction")
plt.show()


```

```python
#Attrition by Department
department_cols = ['Department_Research & Development','Department_Sales']

df[department_cols + ['Attrition_Yes']].groupby(department_cols).mean()

plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Department_Sales', hue='Attrition_Yes')
plt.title("Sales Department Attrition")
plt.show()

```

```python
#Attrition by Job Role
job_role_cols = [col for col in df.columns if col.startswith("JobRole_")]

role_attrition = df[job_role_cols + ['Attrition_Yes']].groupby(job_role_cols).mean()
role_attrition
job_counts = df.groupby('Attrition_Yes')[job_role_cols].sum().T
job_counts.plot(kind='bar', figsize=(10,6))
plt.title("Attrition Count by Job Role")
plt.show()
```

```python
#Attrition vs Overtime
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='OverTime_Yes', hue='Attrition_Yes')
plt.title("Attrition vs Overtime")
plt.show()
```

```python
#Experience Related Factors
plt.figure(figsize=(8,4))
sns.boxplot(data=df, x='Attrition_Yes', y='YearsAtCompany')
plt.title("Years at Company vs Attrition")
plt.show()

#Years in Current Role
plt.figure(figsize=(8,4))
sns.barplot(data=df, x='Attrition_Yes', y='YearsInCurrentRole')
plt.title("Years in Current Role vs Attrition")
plt.show()

#Years Since Last Promotion
plt.figure(figsize=(8,4))
sns.lineplot(data=df, x='Attrition_Yes', y='YearsSinceLastPromotion')
plt.title("Promotion Gap vs Attrition")
plt.show()
```

```python
#Relationship Satisfaction vs Attrition
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='RelationshipSatisfaction', hue='Attrition_Yes')
plt.title("Relationship Satisfaction vs Attrition")
plt.show()
```

```python
#correlation Heatmap
plt.figure(figsize=(16,10))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Most Important Features (Correlation with Attrition)
corr = df.corr()['Attrition_Yes'].sort_values(ascending=False)
print(corr.head(20))
print(corr.tail(20))
```

```python
#Top Predictors Visualization
#for positive
plt.figure(figsize=(6,8))
corr.head(15).plot(kind='barh')
plt.title("Top Positive Attrition Correlations")
plt.show()

#for negative
plt.figure(figsize=(6,8))
corr.tail(15).plot(kind='barh')
plt.title("Top Negative Attrition Correlations")
plt.show()
```
